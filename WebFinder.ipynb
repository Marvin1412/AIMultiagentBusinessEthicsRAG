```python
import requests
import os
from google.colab import drive

# Google Drive einbinden
drive.mount('/content/drive')

# Verzeichnis in Google Drive festlegen
drive_folder = '/content/drive/My Drive/data/KarlHomann'
os.makedirs(drive_folder, exist_ok=True)

# Erweiterte Liste mit zus√§tzlichen Texten von Karl Homann
texts = {
    'The Sense and Limits of the Economic Method in Business Ethics': 'https://www.wcge.org/images/wissenschaft/publikationen/DP_2006-5_Homann_-_The_Sense_and_Limits_of_the_Economic_Method_in_Business_Ethics_o.pdf',
    'Competition and Morality': 'https://www.wcge.org/images/wissenschaft/publikationen/DP_2006-4_Homann_-_Competition_and_Morality_o.pdf',
    'The Scientific Approach of the Wittenberg Center for Global Ethics (WCGE) and the Role of Business': 'https://www.wcge.org/images/wissenschaft/publikationen/DP_2013-01_Homann_The_Scientific_Approach_of_the_Wittenberg_Center_for_Global_Ethics_WCGE.pdf',
    'Die Interdependenz von Zielen und Mitteln': 'https://ageconsearch.umn.edu/record/300254/files/Schmitt_1983_Die_Interdependenz_von_Zielen_und_Mitteln.pdf',
    "There's Life in the Old Dog Yet: The Homo Oeconomicus Model and its Value for the Study of Morality": 'https://www.researchgate.net/publication/345628952_There%27s_Life_in_the_Old_Dog_Yet_The_Homo_Oeconomicus_Model_and_its_Value_for_the_Study_of_Morality'
}

# Herunterladen und Speichern der Texte
for title, url in texts.items():
    response = requests.get(url)
    if response.status_code == 200:
        file_extension = url.split('.')[-1]
        file_path = os.path.join(drive_folder, f"{title}.{file_extension}")
        with open(file_path, 'wb') as file:
            file.write(response.content)
        print(f"'{title}' wurde erfolgreich heruntergeladen und in Google Drive gespeichert.")
    else:
        print(f"Fehler beim Herunterladen von '{title}'. Statuscode: {response.status_code}")

# Best√§tigung der gespeicherten Dateien
print("\nDateien im Ordner:")
print(os.listdir(drive_folder))
```

```python
import os
import requests
import re
from google.colab import drive

# Google Drive einbinden
drive.mount('/content/drive')

# üìÅ Set download folder in Google Drive
drive_folder = '/content/drive/My Drive/data/BusinessEthics'
os.makedirs(drive_folder, exist_ok=True)

# üîç List of Business Ethics PDFs (Manually Curated)
business_ethics_pdfs = {
    "Business Ethics: A Manual for Managing a Responsible Business Enterprise in Emerging Market Economies": "https://legacy.trade.gov/goodgovernance/adobe/bem_manual.pdf",
    "Applying Business Ethics: A guide to help you deal with the many ethical issues that can arise in the course of operating a business": "https://www.failteireland.ie/FailteIreland/media/WebsiteStructure/Documents/2_Develop_Your_Business/1_StartGrow_Your_Business/Applying-Business-Ethics-BT-ABE-C9-0913-4.pdf",
    "Code of Business Ethics and Conduct - American Red Cross": "https://www.redcross.org/content/dam/redcross/about-us/governance/Code-of-Business-Ethics-and-Conduct.pdf",
    "Business Ethics: The Law of Rules": "https://www.hks.harvard.edu/sites/default/files/centers/mrcbg/programs/cri/files/workingpaper_19_michael.pdf",
    "Business Ethics - OpenStax": "https://assets.openstax.org/oscms-prodcms/media/documents/BusinessEthics-OP.pdf",
    "Business Ethics: Ethical Decision Making and Cases": "https://digitalcommons.liberty.edu/cgi/viewcontent.cgi?article=1001&context=busi_fac_pubs",


}
def download_pdfs(pdf_links):
    """
    Downloads PDFs and saves them to Google Drive.
    """
    for title, url in pdf_links.items():
        try:
            response = requests.get(url, timeout=10)
            if response.status_code == 200:
                # üè∑ Clean title for filename
                file_name = re.sub(r"[^\w\s]", "", title)[:50] + ".pdf"
                file_path = os.path.join(drive_folder, file_name)

                # üíæ Save the PDF file
                with open(file_path, "wb") as file:
                    file.write(response.content)

                print(f"‚úÖ '{title}' successfully downloaded and saved.")
            else:
                print(f"‚ö† Error downloading '{title}'. Status code: {response.status_code}")
        except requests.RequestException as e:
            print(f"‚ùå Download error for '{title}': {e}")

# üì• Download and save all Business Ethics PDFs
download_pdfs(business_ethics_pdfs)

# üìÇ Confirm saved files
print("\nüìÅ Saved Files:")
print(os.listdir(drive_folder))

```
